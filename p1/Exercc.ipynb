```python
tarefas = {
    1: {"descricao": "Fazer compras", "prioridade": "média", "concluida": False},
    2: {"descricao": "Estudar Java", "prioridade": "alta", "concluida": False},
    3: {"descricao": "Lavar o civicao do pai", "prioridade": "baixa", "concluida": True}
}

def adicionar_tarefa():
    print("Adicionar nova tarefa:")
    descricao = input("Descrição: ").strip()
    
    while True:
        prioridade = input("Prioridade (baixa/média/alta): ").lower().strip()
        if prioridade in ["baixa", "média", "media", "alta"]:
            prioridade = prioridade.replace("media", "média")  
            break
        print("Prioridade inválida! Use 'baixa', 'média' ou 'alta'")
    
    nova_chave = max(tarefas()) + 1 if tarefas else 1
    tarefas[nova_chave] = {
        "descricao": descricao,
        "prioridade": prioridade,
        "concluida": False
    }
    print(f"Tarefa '{descricao}' adicionada com sucesso!")

def mostrar_tarefas():
    print("Lista de Tarefas:")
    print("-" * 40)
    for id_tarefa, info in tarefas.items():
        status = "✓" if info["concluida"] else " "
        print(f"{id_tarefa}. [{status}] {info['descricao']} ({info['prioridade']})")
    print("-" * 40)

while True:
    print("Menu:")
    print("1. Ver tarefas")
    print("2. Adicionar tarefa")
    print("3. Sair")
    
    opcao = input("Escolha uma opção: ")
    
    if opcao == "1":
        mostrar_tarefas()
    elif opcao == "2":
        adicionar_tarefa()
    elif opcao == "3":
        print("Saindo do sistema...")
        break
    else:
        print("Opção inválida! Tente novamente.")
```

```python
from datetime import datetime

pessoas = { "João": "15/03/1990", "Maria": "22/08/1985", "Pedro": "10/11/1995"
}

data_atual = datetime.now()

print("Idades das pessoas:")
print("-" * 20)

for nome, data_nascimento in pessoas.items():
    data_nascimento = datetime.strptime(data_nascimento, "%d/%m/%Y")
    
    idade = data_atual.year - data_nascimento.year
    
    if (data_atual.month, data_atual.day) < (data_nascimento.month, data_nascimento.day):
        idade -= 1
    
    print(f"{nome}: {idade} anos")

print("-" * 20)
```
